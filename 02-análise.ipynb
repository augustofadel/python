### Exercício

Crie uma função (usando `def`) que recebe como agrumento um array e retorne uma mensagem informando o valor máximo e em qual posição do array ele está.

🛑 Atenção à identação!

```python
def acha_maximo(vetor_maximo):
  maximo = np.max(vetor_maximo)
  posicao = np.argmax(vetor_maximo)
  print('o valor máximo é', maximo, 'e ele está na posição', posicao)
```

```python
teste = np.random.randint(0, 100, 10)
teste
```

```python
acha_maximo(teste)
```

De maneira análoga, crie uma função que, a partir de um array, retorne o valor mínimo e sua posição.

```python
def array_min(a):
  minimo = np.min(a)
  posicao = np.argmin(a)
  print('o valor mínimo é', minimo, 'e ele está na posição', posicao)

array_min(teste)
```

Crie uma função que, além de um array, receba como argumento uma string com dois possíveis valores: `min` ou `max`.

A função deverá retornar posição e valor do elemento mínimo ou máximo conforme esecificado pelo usuário.

```python
def array_min_max(vetor, opcao):
  if opcao == 'minimo':
    minimo = vetor.min()  # np.min(vetor)
    posicao = vetor.argmin()  # np.argmin(vetor)
    print('o valor mínimo é', minimo, 'e ele está na posição', posicao)
  else:
    maximo = vetor.max()  # np.min(vetor)
    posicao = vetor.argmax()  # np.argmin(vetor)
    print('o valor máximo é', maximo, 'e ele está na posição', posicao)

array_min_max(teste, 'maximo')
```

### Combinando dataframes

```python
pd.concat([df1, df2])
```

#### Merge

Permite combinar 2 dataframes por uma ou mais colunas.

```python
np.random.seed(123)

df1 = pd.DataFrame({
    'id': ['0001', '0002', '0004'],
    'nome': ['Joao', 'Maria', 'Joana'],
    'x': randn(3,),
    'y': randn(3,)
})
df2 = pd.DataFrame({
    'id': ['0001', '0002', '0003'],
    'w': randn(3,),
    'z': randn(3,)
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='inner', on='id')
```

```python
pd.merge(df1, df2, how='outer', on='id')
```

#### Join

Similar ao merge, mas usando índices em vez de colunas (mais robusto).

```python
np.random.seed(123)

df1 = pd.DataFrame({
    'nome': ['Joao', 'Maria', 'Joana'],
    'x': randn(3,),
    'y': randn(3,)},
    index=['0001', '0002', '0004']
)
df2 = pd.DataFrame({
    'w': randn(3,),
    'z': randn(3,)},
    index=['0001', '0002', '0003']
)
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how='outer')
```

### Ordenação

```python
df = df1.join(df2, how='outer')
df.sort_values(by='nome')
```

🛑 Observe que o índice foi preservado no exemplo anterior.
