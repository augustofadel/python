### Exerc√≠cio

Crie uma fun√ß√£o (usando `def`) que recebe como agrumento um array e retorne uma mensagem informando o valor m√°ximo e em qual posi√ß√£o do array ele est√°.

üõë Aten√ß√£o √† identa√ß√£o!

```python
def acha_maximo(vetor_maximo):
  maximo = np.max(vetor_maximo)
  posicao = np.argmax(vetor_maximo)
  print('o valor m√°ximo √©', maximo, 'e ele est√° na posi√ß√£o', posicao)
```

```python
teste = np.random.randint(0, 100, 10)
teste
```

```python
acha_maximo(teste)
```

De maneira an√°loga, crie uma fun√ß√£o que, a partir de um array, retorne o valor m√≠nimo e sua posi√ß√£o.

```python
def array_min(a):
  minimo = np.min(a)
  posicao = np.argmin(a)
  print('o valor m√≠nimo √©', minimo, 'e ele est√° na posi√ß√£o', posicao)

array_min(teste)
```

Crie uma fun√ß√£o que, al√©m de um array, receba como argumento uma string com dois poss√≠veis valores: `min` ou `max`.

A fun√ß√£o dever√° retornar posi√ß√£o e valor do elemento m√≠nimo ou m√°ximo conforme esecificado pelo usu√°rio.

```python
def array_min_max(vetor, opcao):
  if opcao == 'minimo':
    minimo = vetor.min()  # np.min(vetor)
    posicao = vetor.argmin()  # np.argmin(vetor)
    print('o valor m√≠nimo √©', minimo, 'e ele est√° na posi√ß√£o', posicao)
  else:
    maximo = vetor.max()  # np.min(vetor)
    posicao = vetor.argmax()  # np.argmin(vetor)
    print('o valor m√°ximo √©', maximo, 'e ele est√° na posi√ß√£o', posicao)

array_min_max(teste, 'maximo')
```

#### Merge

Permite combinar 2 dataframes por uma ou mais colunas.

```python
np.random.seed(123)

df1 = pd.DataFrame({
    'id': ['0001', '0002', '0004'],
    'nome': ['Joao', 'Maria', 'Joana'],
    'x': randn(3,),
    'y': randn(3,)
})
df2 = pd.DataFrame({
    'id': ['0001', '0002', '0003'],
    'w': randn(3,),
    'z': randn(3,)
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='inner', on='id')
```

```python
pd.merge(df1, df2, how='outer', on='id')
```

#### Join

Similar ao merge, mas usando √≠ndices em vez de colunas (mais robusto).

```python
np.random.seed(123)

df1 = pd.DataFrame({
    'nome': ['Joao', 'Maria', 'Joana'],
    'x': randn(3,),
    'y': randn(3,)},
    index=['0001', '0002', '0004']
)
df2 = pd.DataFrame({
    'w': randn(3,),
    'z': randn(3,)},
    index=['0001', '0002', '0003']
)
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how='outer')
```

### Ordena√ß√£o

```python
df = df1.join(df2, how='outer')
df.sort_values(by='nome')
```

üõë Observe que o √≠ndice foi preservado no exemplo anterior.
