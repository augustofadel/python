# M√∫ltiplos registros

Quando uma base de dados armazena hist√≥rico, √© comum a presen√ßa de m√∫ltiplos registros para um mesmo identificador.

Embora fa√ßam refer√™ncia ao mesmo indiv√≠duo (unidade de an√°lise), esses registros n√£o s√£o necessariamente duplicados e podem trazer informa√ß√µes em diferentes situa√ß√µes, como ao longo do tempo, por exemplo.

Identificador como √≠ndice

```python
df = pd.DataFrame(
    data=[
        ['direito', '05/02/2021', 'n√£o'],
        ['direito', '10/02/2021', 'n√£o'],
        ['engenharia', '19/08/2021', 'n√£o'],
        ['enfermagem', '13/02/2022', 'n√£o'],
        ['administra√ß√£o', '01/03/2023', 'sim'],
        ['engenharia', '10/03/2023', 'sim'],
        ['psicologia', '04/09/2023', 'n√£o']
    ],
    index=['000000001-00', '000000002-00', '000000001-00', '000000001-00', '000000002-00', '000000003-00', '000000001-00'],
    columns=['curso', 'data', 'concluido']
)

df
```

Identificador como coluna

```python
df = pd.DataFrame(
    data=[
        ['000000001-00', 'direito', '05/02/2021', 'n√£o'],
        ['000000002-00', 'direito', '10/02/2023', 'n√£o'],
        ['000000001-00', 'engenharia', '19/08/2021', 'n√£o'],
        ['000000001-00', 'enfermagem', '13/02/2022', 'n√£o'],
        ['000000002-00', 'administra√ß√£o', '01/03/2020', 'sim'],
        ['000000003-00', 'engenharia', '10/03/2015', 'sim'],
        ['000000001-00', 'psicologia', '04/09/2023', 'n√£o']
    ],
    columns=['cpf', 'curso', 'data', 'concluido']
)

df
```

```python
df.sort_values(by='cpf')
```

```python
df.sort_values(by='data').drop_duplicates(subset='cpf', keep='last')
```

```python
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')
df.sort_values(by=['cpf', 'data'])
```

```python
df_dedup = df.sort_values(by='data').drop_duplicates(subset='cpf', keep='last')
df_dedup
```

```python
df_dedup.sort_values(by='cpf')
```
