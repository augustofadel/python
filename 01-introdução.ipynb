#### list comprehension

*   equivale ao `for`
*   sintaxe mais sucinta, porÃ©m menos direta
*   execuÃ§Ã£o mais eficiente

```python
base = [1, 2, 3, 4, 5]
```

```python
potencia = []

for i in base:
  potencia.append(i ** 2)

print(potencia)
```

```python
potencia2 = [i ** 2 for i in base]

potencia2
```

```python
nota_alunos = {'Joao': [8, 5], 'Maria': [10, 8], 'Jose': [2, 6]}

print(nota_alunos.keys())
print(nota_alunos.values())
```

```python
for nota in nota_alunos.values():
  print(nota)
  print(sum(nota))
  print(len(nota))
  print(sum(nota) / len(nota))
```

```python
media = []
for nota in nota_alunos.values():
  media.append(sum(nota) / len(nota))

media
```

```python
media2 = [sum(nota) / len(nota) for nota in nota_alunos.values()]

media2
```

#### RepetiÃ§Ã£o: `map`

```python
def quadrado(b):
  return b ** 2

quadrado(3)
```

```python
base = [1, 2, 3, 4, 5]

# potencia(base)
list(map(quadrado, base))
```

#### AnÃ´nimas: `lambda`

```python
def quadrado(b): return b ** 2

quadrado(3)
```

```python
list(map(lambda b: b ** 2, base))
```

SeleÃ§Ã£o de elementos: `filter`

ðŸ›‘ A funÃ§Ã£o aplicada deve executar um teste lÃ³gico, i.e., o retorno da funÃ§Ã£o anÃ´nima deve ser booleano (lÃ³gico): `True` ou `False`.

```python
minhas_notas = [8, 7, 10, 6, 5, 8]

list(filter(lambda nota: nota > 7, minhas_notas))
```

```python
string = ['Estacio', 'Python', 'Curso']

list(filter(lambda x: len(x) > 5, string))
```
